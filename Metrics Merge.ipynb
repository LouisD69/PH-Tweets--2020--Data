```python
import pandas as pd
import csv
import re
import concurrent.futures
import tqdm
import numpy as np
from collections import Counter 
```

```python
tables=[]    
for i in range(1, 4):
    filename = 'tweets' + str(i) + 'utf8_cleaned_metrics.csv'
    tables.append(pd.read_csv(filename).rename(columns={'Unnamed: 0':'emoji'}))
```

```python
full_metrics = pd.concat(tables).groupby(['emoji']).sum().sort_values('total_count', ascending=False)
full_metrics
```

```python
full_metrics.to_csv('tweets_emoji_full_metrics.csv', encoding='utf-8', quoting=csv.QUOTE_ALL)
```
